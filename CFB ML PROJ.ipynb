```python
import pandas as pd
```

```python
matches = pd.read_csv("matches.csv", index_col=0)
```

```python
print(matches["date"].unique())
```

```python
matches = matches[matches["date"].str.contains(r"\d", na=False)]
```

```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["location_code"] = matches["opponent"].str.strip().str[0].apply(lambda x: 0 if x == "@" else 1)
```

```python
matches["clean_opponent"] = matches["opponent"].str.replace(r"^[@+]\s*", "", regex=True)
```

```python
matches["opp_code"] = matches["clean_opponent"].astype("category").cat.codes
```

```python
matches
```

```python
matches["total yards"] = pd.to_numeric(matches["total yards"], downcast="integer")
```

```python
matches["plays"] = pd.to_numeric(matches["plays"], downcast="integer")
```

```python
matches["td"] = pd.to_numeric(matches["td"], downcast="integer")
```

```python
matches["fg"] = pd.to_numeric(matches["fg"], downcast="integer")
```

```python
matches["rush yards"] = pd.to_numeric(matches["rush yards"], downcast="integer")
```

```python
matches["pass yards"] = pd.to_numeric(matches["pass yards"], downcast="integer")
```

```python
matches["points"] = pd.to_numeric(matches["points"], downcast="integer")
```

```python
matches["int."] = pd.to_numeric(matches["int."], downcast="integer")
```

```python
matches["margin"] = pd.to_numeric(matches["margin"], downcast="integer")
```

```python
matches["yards/play"] = pd.to_numeric(matches["yards/play"], downcast="float")
```

```python
matches["target"] = matches["result"].str.startswith("W").astype("int")
```

```python
matches["score"] = matches["result"].str.split().str[-1]
```

```python
matches["opp_points"] = matches["score"].str.split("-").str[-1].astype(int)
```

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Akron")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["total yards", "td", "fg", "margin", "opp_points" , "yards/play", "points"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < "2024-11-01"]
    test = data[data["date"] >= "2024-11-01"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
combined = combined.merge(matches_rolling[["date", "team", "clean_opponent", "result"]], left_index=True, right_index=True)
```

```python
merged = combined.merge(combined, left_on=["date", "team"], right_on=["date", "clean_opponent"])
```
